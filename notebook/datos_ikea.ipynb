```python
import pandas as pd
```

```python
df_ikea = pd.read_csv('ikea.csv')
```

```python
# Rename to df_ikea and drop the specified columns
df_ikea = df_ikea.drop(columns=[
    'plp-product__image-link href', 'plp-image src', 'image src', 
    'plp-commercial-message__title', 'notranslate'
])


```
