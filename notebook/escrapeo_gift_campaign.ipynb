```python
import unidecode
```

```python
lista_productos=[]
lista_precios=[]
for i in range(3):
    url = f'https://www.giftcampaign.es/articulos-publicitarios-cocina/fiambreras-personalizadas.html?p={i+1}'

    response = requests.get(url)

    soup = BeautifulSoup(response.text, 'html.parser')


    contenedor_nombres= soup.findAll('h3', 'product-name product-name-home')
    contenedor_precios=soup.findAll('div', 'price-box')
    for producto in contenedor_nombres:
        lista_productos.append(producto.text.strip())

    for precio in contenedor_precios:
        lista_precios.append(precio.text.strip())

```

```python
lista_productos
```

```python
df_giftcampaign = pd.DataFrame({
    'productos': lista_productos,
    'precio': lista_precios
})
```

```python
df_giftcampaign.head()
```

```python
df_giftcampaign['precio'] = df_giftcampaign['precio'].str.replace("Precio desde:\n", "", regex=False)
df_giftcampaign['precio'] = df_giftcampaign['precio'].str.replace("€", "", regex=False)

df_giftcampaign.head()
```

```python
df_giftcampaign['precio'] = df_giftcampaign['precio'].str.replace(',', '.').astype(float)
```

```python
df_giftcampaign.head()
```

```python
df_giftcampaign.rename(columns={'productos': 'producto'}, inplace=True)

df_giftcampaign
```

```python
df_giftcampaign = df_giftcampaign.applymap(lambda x: unidecode.unidecode(x) if isinstance(x, str) else x)
```
