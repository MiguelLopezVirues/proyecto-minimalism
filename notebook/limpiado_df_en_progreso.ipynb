```python
cadenas_a_borrar = ["*", "-", "algodon organico", "algodon reciclado", "lana merino reciclada"]

# Función para eliminar las cadenas específicadas
def borrar_cadenas(text, strings):
    for s in strings:
        text = text.replace(s, "")
    return text.strip()

# Aplicar la función a la columna 'nombre'
df_precios_minimalism['nombre'] = df_precios_minimalism['nombre'].apply(lambda x: borrar_cadenas(x, cadenas_a_borrar))
```

```python
import re

# Definir una función para extraer los datos específicos
def extract_impacto(text):
    if isinstance(text, str):
        # Buscar los patrones de kg de CO₂, litros de agua y kWh de energía
        co2 = re.search(r'(\d*,?\d*)\s*kg de emisiones de co2 evitados', text)
        agua = re.search(r'\d+\.?\d*\s*litros de agua ahorrados', text)
        energia = re.search(r'\d+\.?\d*\s*kwh de energia ahorrados', text)
        
        # Crear una lista con los valores encontrados
        results = [match.group() for match in [co2, agua, energia] if match]
        return ', '.join(results)
    return text

# Aplicar la función a la columna 'impacto'
df_precios_minimalism['impacto'] = df_precios_minimalism['impacto'].apply(extract_impacto)






```
